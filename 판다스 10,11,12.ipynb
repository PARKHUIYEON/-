```python
#선 그래프 2개 동시에 그리기

x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, label='1+sin', color='blue', alpha=0.3, marker='o') #marker로 찍히는 점 모양 지정 가능
plt.plot(x, y_2, label='1+cos', color='red', alpha=0.7, linestyle='-.') #선모양 지정 가능

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)
plt.title('sin and cos graph', fontsize=18)

plt.grid()
plt.legend()

plt.show()
```

```python
#히스토그램 여러 개 그리기

N = 100000
bins = 30

x = np.random.randn(N)

fig, axs = plt.subplots(1, 3, 
                        sharey=True, 
                        tight_layout=True
                       ) #그릴 곳 만들고

fig.set_size_inches(12, 5)# 그림 크기 지정

axs[0].hist(x, bins=bins) #0번 에 넣을 그래프
axs[1].hist(x, bins=bins*2) #1번에 넣을 그래프
axs[2].hist(x, bins=bins*4) #2번에 넣을 그래프

plt.show()
```

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axs = plt.subplots(1, 2, 
                        tight_layout=True
                       )
fig.set_size_inches(9, 3)

#density=True 이면 밀도함수(Y축이 상대도수 값)
axs[0].hist(x, bins=bins, density=True, cumulative=True) #cumulative : 누적으로 그릴지 아닐지 옵션
axs[1].hist(x, bins=bins, density=True)

plt.show()
```

파이 차트 
- explode : 튀어나온 비율
- autopct : 퍼센트 자동 표기
- shadow : 그림자 유무
- startangle : 파이를 그리기 시작할 각도

```python
labels = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'Oppo', 'Etc']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]
explode = (0.3, 0, 0, 0, 0, 0)

patches, texts, autotexts = plt.pie(sizes, 
                                    explode=explode, # 삼성은 0.3튀어나오고 다른건 다 0 
                                    labels=labels,  
                                    autopct='%1.1f%%',
                                    shadow=True, 
                                    startangle=90)

plt.title('Smartphone pie', fontsize=15)

for t in texts:
    t.set_fontsize(12)
    t.set_color('gray')

for t in autotexts:
    t.set_color("white")
    t.set_fontsize(18)

plt.show()
```

```python
from IPython.display import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

plt.rcParams['axes.unicode_minus']=False

plt.rcParams["figure.figsize"] = (10, 8)
```

```python
import seaborn as sns

tips=sns.load_dataset('tips')#데이터셋 가져오기
tips
```

```python
sns.violinplot(x="day", y="total_bill", data=tips) #바이올린 플랏
#day와 total bill  사이의 관계를 알아서 나타내줌
plt.show()
```

```python
sns.countplot(x=tips['day']) #막대그래프, day에 관해 알아서 카테고리별로 나타내줌
plt.show()
```

```python
#relplot은 x,y 간의 상관관계를 산점도 형식으로 출력
sns.relplot(x='tip', y='total_bill', data=tips)
plt.show()
```

```python
#relplot에 선이 함께 출력(상관관계를 더 직관적으로 확인)
sns.lmplot(x='tip', y='total_bill', data=tips)
plt.show()
```

```python
plt.title('heatmap')
sns.heatmap(tips.corr(), annot=True, linewidths=1)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
fig.set_size_inches(12, 6)

#맷플롯립보다 시본이 더 깔끔하고 보기 쉽게 나타내줌
#검은 선은 신뢰구간 95%

axes[0].bar(tips['day'], tips['total_bill'])
axes[0].set_title('matplotlib')
sns.barplot(x="day", y="total_bill", data=tips, palette='Set1', ax=axes[1])
axes[1].set_title('seaborn')
plt.show()
```

```python
sns.catplot(x="sex", y="total_bill",
            hue="smoker", #hue로 smoker의 카테고리 값에 따라 알아서 분류
            data=tips, 
            col = 'time', # time의 값들에 따라 알아서 열별로 막대 그래프 분류
            kind="bar") # 성별과 total_bill 상관관계를 막대그래프로 보여줌
plt.show()
```
