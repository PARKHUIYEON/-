```python
df = sns.load_dataset("titanic") #연습문제
df.head()
```

```python
condition = (df['age'] >= 30) #연습문제

condition2 = (df['who'] == 'man')

dff = df.loc[condition & condition2]
dff.sort_values('fare', ascending=False).head(10)
```

```python
condition3 = (df['age'] >= 20) #연습문제
condition4 = (df['age'] < 40)
condition5 = (df['pclass'] == 1)
condition6 = (df['pclass'] == 2)

ddff = df.loc[condition3 & condition4 & (condition5 | condition6)]
ddff
```

```python
ddff.loc[:, ['survived', 'pclass', 'age', 'fare']] #연습문제
```

```python
#df.where(T/F판단할 수 있는 식, T/F 판단 못 할때 출력할 값)

df['fare'].where(df['fare'] < 20, 0).tail(10) #fare만 확인하고 fare 값이 20보다 작으면 값을, 아니면 0.
```

```python
df.where(df['fare'] < 10) #범위를 전체 데이터 프레임으로 적용하고 fare이 10보다 작은 값, 아니면 NaN.
```
