```python
df1 = pd.DataFrame({
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
df2
```

```python
pd.merge(df1, df2) #merge > 말그대로 병합 , 공통인 칼럼은 그냥 하나로, 따로 있는 칼럼은 앞에 있는 거 부터 차례로 붙임. 공통아닌 건 제외됨
```

## how{left, right, outer, inner}
디폴트는 inner(교집합 값)

```python
#left하면 왼쪽 데이터(df1)에서 공통아닌 것도 가져와서 붙임
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right') #right는 오른쪽 데이터에서 
```

```python
pd.merge(df1, df2, how='outer') #outer는 합집합 값
```

```python
df1 = pd.DataFrame({
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
df2
```

#### 데이터 값은 같은데 칼럼 이름이 다를 경우


```python
#left_on 왼쪽 데이터 의 '이름'  칼럼, right_on 오른쪽 데이터 의 '고객명' 칼럼 이 같은 거 라고 말해줌

pd.merge(df1, df2, left_on='이름', right_on='고객명')
```

```python
data = np.arange(100, 201)
plt.subplot(2, 1, 1) #subplot(행, 열, 2개 칸 중 1번 칸), 콤마 는 생략 가능
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(2, 1, 2)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100, 201)
plt.subplot(1, 3, 1) # 1행 3열로 나눈 다음 그 중 1번칸
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(1, 3, 2)
plt.plot(data2)

data3 = np.arange(300, 401)
plt.subplot(1, 3, 3)
plt.plot(data3)

plt.show()
```

```python
data = np.arange(1, 51)

fig, ax = plt.subplots(2, 3) #subplots 는 그래프를 그릴 판(?) 을 먼저 만들어놓고 그 위에 plot으로 그래프 레이어를 얹는 방식

ax[0,0].plot(data) #0행 0열 그래프는 데이터 그대로 
ax[0,1].plot(data*data) # 0행 1열 그래프는 데이터와 데이터를 곱한 그래프
ax[0,2].plot(data**3) # ...
ax[1,0].plot(data%10)
ax[1,1].plot(-data)
ax[1,2].plot(data//20)

plt.tight_layout()
plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10)*2)
plt.plot(np.arange(10), np.arange(10)**2)
plt.plot(np.arange(10), np.log(np.arange(10)))

plt.title('이것은 타이틀', fontsize=50)

plt.xlabel('X축', fontsize=50)
plt.ylabel('Y축', fontsize=50)

plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.legend(['10 * 2', '10 ** 2', 'log'], fontsize=15)

plt.xlim(0, 5) #xlim으로 x 값의 표시(?) 범위를 제한
plt.ylim(0.5, 10) # ylim으로 y 값의 표시 범위를 제한

plt.show()
```

```python
plt.plot(np.arange(10), np.arange(10)*2, marker='o', markersize=5, linestyle='', color='b', alpha=0.1)
#marker로 찍힐 모양 선택 가능, makersize 지정 가능
plt.plot(np.arange(10), np.arange(10)*2 - 10, marker='v', markersize=10, linestyle='-', color='c', alpha=0.3)
# 점과 점 사이의 선이 그어지는 모양도 지정 가능
plt.plot(np.arange(10), np.arange(10)*2 - 20, marker='+', markersize=15, linestyle='--', color='y', alpha=0.6)
# 색도 지정 가능, 원래 있는 거 쓰거나 색 번호로 지정하거나
plt.plot(np.arange(10), np.arange(10)*2 - 30, marker='*', markersize=20, linestyle='-.', color='r', alpha=1.0)
# alpha로 투명도 0~1 설정

plt.title('그래프 설정', fontsize=50)

plt.xlabel('X축', fontsize=50)
plt.ylabel('Y축', fontsize=50)

plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.annotate('여기', xy=(3, -20), xytext=(3,-25),arrowprops=dict(facecolor='black', shrink=0.05))
# 특정 부분 표시 가능, 표시 위치와 텍스트가 들어갈 위치 , arrowprops로 화살표 모양 설정도 가능

plt.grid() #그리드 표시 추가 가능

plt.show()
```
