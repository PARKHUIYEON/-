```python
grouped = df.groupby('Team') # Team을 기준으로 묶기만
grouped.get_group('Riders') # 묶은 중 Riders 값만 반환
```

```python
# agg() : 통계 함수

grouped.agg(min) # 묶은 기준('Team') 으로 숫자 값들에 대해 최솟값 반환.
```

```python
grouped.agg(np.mean) # 평균
```

```python
# transform() : 변환 함수

grouped.transform(max) # 묶은 기준에 따라 각 데이터의 최댓값으로 전부 반환
```

```python
# filter() : 특정 조건 만족하는 데이터 검색

df.groupby('Team').filter(lambda x:len(x) >= 3) # 갯수가 3개 이상인 데이터만 필터링
```

```python
df.groupby('Team').filter(lambda x:x['Points'].max() > 800) 
# 팀을 기준으로 묶은 데이터에서 포인트 최댓값이 800을 넘는 팀만 추출하는데 Kings를 제외하고는 전부 최댓값이 800이상임
```

```python
import pandas as pd # pandas 모듈 호출
raw_data = { 'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
             'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_left = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_left
```

```python
raw_data = {
    'subject_id': ['4', '5', '6', '7', '8'],
    'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_right = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_right
```

```python
# merge() : 조인 함수
# left = 왼쪽에 놓을 데이터
# right = 오른쪽에 놓을 데이터
# how = 'inner' : 내부조인 --> 모두 존재하는 열 만 
#       'outer' : 외부조인 --> 모든 여
#       'left' : 왼쪽조인 --> 왼쪽 데이터에 있는 열 기주
#       'right' : 오른쪽조인 --> 오른쪽 데이터에 있는 열 기준
# on = 병합의 기준이 되는 열
# left_on, right_on = 두 데이터의 합칠 열이름이 다를 때

pd.merge(left=df_left, right=df_right, how='inner', on='subject_id')
```

```python
def find_nearest_value(X, target_value):
  result = X[np.argmin(np.abs(X - target_value))]
  return result
```

```python
find_nearest_value(np.array([1,2,3,4,5,6,7]), 4.3)
```
