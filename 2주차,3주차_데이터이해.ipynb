```python
test_array = np.arange(1,13).reshape(3,4)
test_array.sum(axis=0)
```

```python
test_array.mean() #평균
```

```python
test_array.std() #표준편차
```

```python
test_array.std(axis=0)
```

```python
test_array > 5 # 비교연산은결과 T/F
```

```python
np.where(test_array>5) # 결과 : 조건 만족하는 행 값 배열, 열 값 배열.
```

```python
np.where(test_array>5, 10, 20) # 조건 만족하면 10 아니면 20
```

```python
# np.argsort(), np.argmax(), np.argmin() 모두 인덱스를 결과로 반환

x = test_array
np.argsort(x)
```

```python
np.argmax(x)
```

```python
np.argmin(x)
```

```python
x = np.array([4,6,7,3,2])
cond = x > 3
x[cond] # 조건을 만족하는 데이터만 반환
```
