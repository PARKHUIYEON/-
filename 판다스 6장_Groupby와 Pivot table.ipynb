```python
from IPython.display import Image
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df['who'].value_counts()
```

```python
def transform_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'
```

```python
df['who'].apply(transform_who) #apply로 'who'의 값들을 남자, 여자, 아이로 적용
```

```python
df['who'].apply(transform_who).value_counts()
```

```python
def transform_who(x):
    return x['fare'] / x['age']
```

```python
df.apply(transform_who, axis=1) # 왜 구하는지 모르겠음
```

```python
#pipe() : 데이터프레임이나 시리즈에 함수를 연계할 때

df['age'].pipe(lambda x: x+2).pipe(lambda x:x.fillna(x.mean())) 
#--> nan값은 평균으로 채우고, 아닌 것들은 +2
```

```python
df1 = df.pivot_table(index='who', columns='pclass', values='survived', aggfunc=['sum','mean']).copy()
#aggfunc : 통계 함수.?
# who를 인덱스로 pclass를 컬럼으로 정리한 다음 걔들의 컬럼별 합이랑 평균 값을 나타냄 

df1
```

```python
#droplevel()은 멀티 인덱스를 싱글 인덱스 컬럼으로 변환

df1.columns = df1.columns.droplevel()
df1
```

```python
#unstack은 stack반대

df1 = df.groupby(['pclass','who','embarked'])[['age']].mean().copy() 
# pclass , who, embarked로 인덱스를 묶고 age의 평균 값 나타냄
df1
```

## get_dummies는 더미(?) 변수를 생성하는데 수치형 컬럼은 제외하고 생성함

```python
df.iloc[:, :3] #survived랑 pclass는 수치형이니까
```

```python
pd.get_dummies(df.iloc[:, :3]) #더미를 생성하면 수치형이 아닌 sex만 나뉨
```
