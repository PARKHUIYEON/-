## pd.cut으로 연속형 값을 같은 길이로 나눈 구간으로 카테고리화 함
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!pd.cut 하나도 이해안됨

```python
bins = [0, 200, 400, df2['운동량'].max()]
pd.cut(df2['운동량'], bins, right=False) # 결과가 뭘 뜻하는가
```

```python
labels = ['운동부족','보통','많음']
```

```python
pd.cut(df2['운동량'], bins, labels=labels, right=False)
```

###  pd.qcut() = 데이터의 분포를 최대한 비슷하게 유지하는 구간 분할



```python
df2['운동량_qcut'] = pd.qcut(df2['운동량'], q=10)
```

```python
df2['운동량_qcut'].value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1] # 임의 범위 조정
```

```python
pd.qcut(df2['운동량'], qcut_bins)
```

```python
qcut_labels = ['적음','보통','많음'] #라벨 지정(범위수 -1)
```

```python
pd.qcut(df2['운동량'], qcut_bins, labels=qcut_labels).value_counts()
```
