```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
df1.tail()
```

```python
df1['age'].fillna(700).tail() #'age'가 NaN인 값을 fillna() 함수로 700으로 바꿈
```

```python
df1['age'] = df1['age'].fillna(700)
df1.tail()
```

```python
df1.fillna(method='bfill')
#backfill/bfill : 뒤에 위치한 행의 값으로 앞의 값을 채움
#pad/ffill : 앞에 위치한 행의 값으로 뒤의 값을 채움
#None : 아무 변화 X (생략가능)
```

```python
sample = df.iloc[:10, :5].copy() #10행, 5열까지 카피
sample.loc[np.arange(1, 9, 2), 'age'] = np.nan # age의 1,3,5,7 값 NaN
sample.iloc[4, 3] = np.nan # 4행 3열(age) 값 NaN 
sample
```

```python
sample['age'].fillna(method='bfill') # 26, 54, 27 이 앞의 NaN에 입력됨
```

```python
sample['age'].fillna(method='ffill') # 22, 26, 54 가 뒤의 NaN에 입력됨
```

```python
df1['deck'].fillna('A') # 카테고리 형에서는 fillna()에 NaN에 채워줄 값(이미 있는 카테고리(위의 자료는 A~G까지)로) 입력
```

```python
# 없는 카테고리는 add_categories()

df1['deck'].cat.add_categories('No Data').fillna('No Data') #cat은 category 지정자
#No Data를 추가 한 다음 결측치를 No Data로 바꿈
```

```python
#연습문제 4-5 

df = sns.load_dataset('titanic')
df.head()
```

```python
cond1 = (df['sex'] == 'male')
male = df.loc[cond1, 'age'].mean()
male
```

```python
cond2 = (df['sex'] == 'female')
female = df.loc[cond2, 'age'].mean()
female
```

```python
df.loc[cond1, 'age'] = male
```

```python
df.loc[cond2, 'age'] = female
```

```python
assert (df['age'].isnull().sum() == 0)
df['age'].mean()
```

```python
df1['deck'].mode()
```

```python
df1['deck'].fillna(df1['deck'].mode()[0]).tail() # deck의 NaN 값을 mode() 함수로 최빈값인 C로 채운다.
```

```python
df1 = df.copy()
df1.tail()
```

```python
df1.dropna() #1개라도 NaN이 있는 행을 지움
```

```python
df1.dropna(how='all') # 기본 설정은 how=any, how='all' 은 행의 모든 값이 NaN이면 행을 지움
```

```python
df1.isnull().sum()
```

```python
df1.dropna(axis=1) #NaN 값이 존재하는 열을 지움
```

```python
df1.dropna(thresh=14) # thresh=n 을 이용하여 한 행에 최소 n개 이상의 값을 가지는 행만 남겨둔다.
```

```python
df1.notnull().sum()
```

```python
df1.dropna(subset=['age']) #subset=[열 속성] 을 지정하여 특정 칼럼에서 NaN이면 그 행을 지움
```

```python
df1.dropna(subset=['age','deck']) # 열 속성을 여러 개 지정 하면 하나라도 NaN인 열의 행을 지움
```

```python
df1.dropna(subset=['age','deck'], how='all') #how=all 을 추가하면 속성이 모두 NaN이여야 지움
```
